```python
import re

index_pattern = re.compile(r'^\S+\(\d+\)$')
units_pattern = re.compile(r'\s\(.+\)$')
print units_pattern.search('stuff (s)')

print units_pattern.search('stuff(s)')


print index_pattern.search('stuff(1)')
print index_pattern.search('asdf stuff(1)')
print index_pattern.search('stuff (1)')


rxn_sens_pattern = re.compile('^dln\[\S+\]\/dln\[k\d+\]:\s\S+$')
thermo_sens_pattern = re.compile('^dln\[\S+\]\/dG\[\S+\]$')
print 'checking rxnstring'
rxnstring = 'dln[asdlfkj(1)]/dln[k2039]: blah(1)+S(3)=CH(45)'
print rxn_sens_pattern.search(rxnstring)
thermostring = 'dln[alskdj(1)]/dG[stuff(2)]'
print rxn_sens_pattern.search(thermostring)
print thermo_sens_pattern.search(thermostring)
```
